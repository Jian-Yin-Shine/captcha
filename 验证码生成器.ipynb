- 制作验证码数据集

```
import os
import random
import string
import tqdm
```

```
train_dir = 'train'
test_dir = 'test'
```

```
string.digits
```

```
string.ascii_letters
```

```
all_chars = string.digits + string.ascii_letters
```

```
all_chars
```

```
for i in tqdm.trange(4, 9):
  for j in range(2000):
    code = ''.join(random.choices(all_chars, k=i))
    img = generator.generate_image(code)
    
    file_name = os.path.join(train_dir, code +'.jpg')
    img.save(file_name)
```

```
for i in tqdm.trange(4, 9):
  for j in range(200):
    code = ''.join(random.choices(all_chars, k=i))
    img = generator.generate_image(code)
    
    file_name = os.path.join(test_dir, code +'.jpg')
    img.save(file_name)
```
